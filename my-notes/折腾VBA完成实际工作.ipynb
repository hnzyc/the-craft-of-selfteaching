# 20191203折腾VBA完成实际工作
因为学习了笑来老师的前面几章，正好工作上有一个模式，每天不断地做新的项目，有时候一天就会有五六单业务，业务模式已经非常成熟，只是每一个项目都有一套十份合同文本，每次根据项目的资产情况，需要将大约50个变量信息填入这十份合同的不同的地方。

正常情况下，一个熟练工，专心不受任何外界干扰的情况下，修改完成一整套合同大约需要三四个小时，还不保证没有错误，毕竟人的精力是有限的，总是可能出现漏填的情形，如果平时项目时间周期比较长，倒也还好，多个人复核避免错误，但是现在已经增加到每周十来单业务，涉及多家银行，每个项目的人员对项目的熟悉程度不同，又极度复杂，项目对各个变量的准确度要求极高，如果出现一处错误，就导致两家银行三家机构重新走审批流程，这在银行是不可接受的。

这个周末，我就思考，其实所谓的修改合同，就是将整套合同的不同内容，按照不同项目的具体变量，填入到对应的地方去，本质上就是一个批量查找替换。相通了这一点，这周末就开工写代码了。

首先，因为工作中大家都是使用word和pdf作为工作交流的基本格式，他们并不懂md格式是什么，所以，首选还是用VBA来解决实际问题，毕竟VBA对office的支持还是很好的，python需要调用相关的库，我暂时还不具备足够的知识，但是有一点是确定的，掌握了函数就是实现特定需求的功能这一个特点，我就是需要写出一个批量查找替换的代码，然后遍历文件夹所有文件，做个循环就好。

其次，笑来老师也说了，反正以后也会顺带学习下其他语言的，就从VBA开始学习啦，哈哈，而且，我也开始使用git来推送到自己的GitHub账号，保存自己的学习经过，见证自己的学习历程。

第一次写出来的代码，就不在这里贴了，实在不叫代码，因为我写不出循环来遍历文件夹。后来折腾了一整天，终于搞定了，参考了excelhome论坛几个牛人的代码框架最终实现了合同的修改。

现在团队把我给膜拜的不行了，原来三四个小时的工作，现在大家只需要点一下，包括原始数据的输入，前后也用不到5分钟，而且保证没有错误，下周一周我们十单业务，本来大家决定全程不睡觉来完成的，现在可以轻松自如的应付了。

啥也不说啦，继续学习！！！
